## Example Usage
This notebook gives a short overview of the command line tool for chord sequence detection

```python
import matplotlib.pyplot as plt
import sys
sys.path.append("../ChordDetection")
import ChordDetection
```

Initialize an instance of DetectChords class...

```python
DC = ChordDetection.DetectChords()
```

Initialize Deep Learning pipeline...

```python
DC.build_cnn_extractor()
DC.build_crf()
```

Initialize visualization parameters...

```python
DC.initialize_chord_axis()
```

Now, set *song_path* to the full path of the song that you would like to transcribe.\
For this example, I am using a song that is already included in data folder.

```python
song_path = '../data/U2 - With Or Without You - Remastered.mp3'
```

We need to convert the song in wav format if it is not already! If it is in wav format, you can skip this step.

```python
DC.mp3_to_wav(song_path)
```

Finally, we are ready to make chord sequence predictions! Next cell's runtime can from 1-3 minutes, depending on your GPU.

```python
predictions_shape = DC.predict_seq()
```

```python
print(f'Songs ends on {predictions_shape[1]//10}th second. Keep this in mind when setting time interval to visualize!')
```

For visualizing chord sequence predictions, I recommend setting figure parameters as follows (feel free to experiment with these since depending on the length of your interval of interest, these parameters should be adjusted for better visual experience).

```python
plt.rcParams.update({'font.size': 20, 'xtick.major.pad': 15, 
                     'ytick.major.pad': 40, 'axes.titlepad': 15,
                     'xtick.bottom': False, 'ytick.left': False, 'figure.figsize': (70, 25)})
```

To see what chords to play in a certain time interval, use method below, where start and end are the beginning and end of the interval in seconds.\
Do not make the first argument greater or equal to the second one - the method does not handle this!

```python
DC.visualize(10, 20)
```

```python
DC.visualize(80, 90)
```

```python
DC.visualize(150, 155)
```

```python
DC.visualize(150, 151)
```
